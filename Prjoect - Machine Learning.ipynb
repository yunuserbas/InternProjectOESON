# Training and Testing Data

```python
y = database['Price']
X = database[['SSD_Capacity', 'Inch','Processor_Speed', 'Postage_Price']]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=85)
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
# The coefficients
print('Coefficients: \n', lm.coef_)
```

```python
predictions = lm.predict( X_test)

plt.scatter(y_test,predictions, color='violet')
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python

# calculate these metrics by hand!
from sklearn import metrics

print('RMSE:', round( np.sqrt(metrics.mean_squared_error(y_test, predictions)),2) )
```

**Predicting Test Data**

```python
sns.distplot((y_test-predictions),bins=50, color='blue');
```
